```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
alpha=0.07
B=np.exp(np.linspace(np.log(3e-2),np.log(200),200))
f=(B/(1.+B) + alpha/(1.+B));
```

```python
plt.figure(figsize=(6,4))
plt.semilogx(B,f,'b-')
plt.semilogx(B,alpha+0.0*B,'k--',linewidth=0.5)
plt.ylim(0., 1)
plt.xlim(0.03, 100)
plt.xlabel('B',fontsize=14)
plt.ylabel('f(B)',fontsize=14)
plt.tick_params(labelsize=12)
plt.savefig('f_of_B', dpi=600, bbox_inches='tight')
plt.show()
```

```python
dfdB=(1.-alpha)/(B+1.)**2.
ftilde=B*dfdB/f
dfdB025=(1.-alpha)/(0.25+1.)**2.
dfdB1=(1.-alpha)/(1.+1.)**2.
f025=(0.25/(1.+0.25) + alpha/(1.+0.25));
f1=(1./(1.+1.) + alpha/(1.+1.));
ftilde025=0.25*dfdB025/f025
ftilde1=1.*dfdB1/f1
plt.figure(figsize=(6,4))
plt.semilogx(B,ftilde,'b-')
plt.semilogx([0.25,0.25],[0.,1.0],'k--',linewidth=0.5)
plt.semilogx([1.0,1.0],[0.,1.0],'k--',linewidth=0.5)
plt.semilogx([0.01,100],[ftilde025,ftilde025],'k--',linewidth=0.5)
plt.semilogx([0.01,100],[ftilde1,ftilde1],'k--',linewidth=0.5)
plt.ylim(0., 0.7)
plt.xlim(0.03, 100)
plt.xlabel('B',fontsize=14)
plt.ylabel(r'$\tilde{f}(B)$',fontsize=14)
plt.tick_params(labelsize=12)
plt.savefig('f_of_B', dpi=600, bbox_inches='tight')
plt.show()
```
